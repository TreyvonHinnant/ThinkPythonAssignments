```python
import csv
from datetime import datetime
import os

FILE_NAME = 'harvest_log.csv'

def get_valid_number(prompt):
    while True:
        try:
            return float(input(prompt))
        except ValueError:
            print("❌ Invalid input. Please enter a numerical.")

def record_harvest():
    crop_name = input("Enter vegetable name: ").capitalize()
    amount = get_valid_number("Enter quantity (numerical): ")

    with open(FILE_NAME, 'a', newline='') as file:
        writer = csv.writer(file)
        writer.writerow([datetime.now().strftime("%Y-%m-%d"), crop_name, amount])
    print(f"✅Recorded {amount} of {crop_name}.")

def view_total_harvest():
    if not os.path.exists(FILE_NAME):
        print("No data recorded yet.")
        return

    totals = {}
    with open(FILE_NAME, 'r') as file:
        reader = csv.DictReader(file)
        for row in reader:
            crop = row['Crop']
            try:
                amount = float(row['Amount'])
                totals[crop] = totals.get(crop, 0) + amount
            except ValueError:
                continue

    print("\n--- Season Totals ---")
    for crop, total in totals.items():
        print(f"{crop}: {total}")
    print("---------------------\n")

if __name__ == "__main__":
    while True:
        choice = input("1. Add Harvest\n2. View Totals\n3. Exit\nChoice: ")
        if choice == '1': record_harvest()
        elif choice == '2': view_total_harvest()
        elif choice == '3': break
        else: print("Invalid choice.")
```

```python
import csv
from datetime import datetime

# Feature 1: Data Entry & File Persistance
def record_harvest():
    crop_name = input("Enter vegetable name (e.g., Tomato): ")
    amount = input("Enter quantity/weight (e.g., 2kg): ")
    date = datetime.now().strftime("%Y-%m-%d")

    # Save data permanetly to a CSV file
    with open('harvest_log.csv', 'a', newline='') as file:
        writer = csv.writer(file)
        writer.writerow([date, crop_name, amount])

    print(f"Recorded {amount} of {crop_name} on {date}.")

if __name__ == "__main__":
    print("--- GardenHarvest Tracker v1.0 ---")
    record_harvest()
```

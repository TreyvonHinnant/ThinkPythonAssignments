```python
import csv
from datetime import datetime
import os

FILE_NAME = 'harvest_log.csv'

if not os.path.exists(FILE_NAME):
    with open(FILE_NAME, 'w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(["Date", "Crop", "Amount"])

def record_harvest():
    crop_name = input("Enter vegetable name: ").capitalize()
    amount = input("Enter quantity (number only): ")

    with open(FILE_NAME, 'a', newline='') as file:
        writer = csv.writer(file)
        writer.writerow([datetime.now().strftime("%Y-%m-%d"), crop_name, amount])
    print(f"✅Recorded {amount} of {crop_name}.")

def view_total_harvest():
    if not os.path.exists(FILE_NAME): return

    totals = {}
    with open(FILE_NAME, 'r') as file:
        reader = csv.DictReader(file)
        for row in reader:
            crop = row['Crop']
            amount = float(row['Amount'])
            totals[crop] = totals.get(crop, 0) + amount

    print("\n--- Season Totals ---")
    for crop, total in totals.items():
        print(f"{crop}: {total}")
    print("---------------------\n")

if __name__ == "__main__":
    while True:
        choice = input("1. Add Harvest\n2. View Totals\n3. Exit\nChoice: ")
        if choice == '1': record_harvest()
        elif choice == '2': view_total_harvest()
        else: break
```

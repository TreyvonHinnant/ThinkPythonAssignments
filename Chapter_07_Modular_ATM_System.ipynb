```python
def verify_pin():
  """Repeatedly ask for a PIN until "1234" is entered."""
  while True:
      pin = input ("Enter PIN: ")
      if pin == "1234":
          print("PIN accepted.")
          break
      else:
          print("Incorrect PIN. Try again.")



# Call the function
verify_pin()

```

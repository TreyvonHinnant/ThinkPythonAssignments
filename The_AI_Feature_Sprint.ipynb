```python
import json
import os

FILE_NAME = "tasks.json"
HISTORY_FILE = "history.txt"

def load_tasks():
    if not os.path.exists(FILE_NAME):
        return []
    with open(FILE_NAME, "r") as f:
        return json.load(f)

def save_tasks(tasks):
    with open(FILE_NAME, "w") as f:
        json.dump(tasks, f, indent=4)

def add_task(title, priority):
    tasks = load_tasks()
    tasks.append({"title": title, "priority": priority, "completed": False})
    save_tasks(tasks)
    print(f"Task '{title}' added with priority {priority}.")

def view_tasks():
    tasks = load_tasks()
    if not tasks:
        print("No tasks found.")
        return
    for i, task in enumerate(tasks, 1):
        status = "✅" if task["completed"] else "❌"
        print(f"{i}. [{task['priority']}] {task['title']} {status}")

def complete_task(task_index):
    tasks = load_tasks()
    if 0 <= task_index < len(tasks):
        task = tasks.pop(task_index)
        with open(HISTORY_FILE, "a") as f:
            f.write(f"{task['title']}\n")
        save_tasks(tasks)
        print(f"Completed: {task['title']}")

def what_is_next():
    tasks = load_tasks()
    p1_tasks = [t for t in tasks if t['priority'] == 1]

    if p1_tasks:
        print(f"👉 Next High Priority Task: {p1_tasks[0]['title']}")
    else:
        print("✨You're all caught up! Time for a break! ✨")

def main():
    while True:
        print("\n1. Add Task\n2. View Tasks\n3. Exit")
        choice = input("Select option: ")
        if choice == "1":
            title = input("Task Name: ")
            priority = int(input("Priority (1-High, 2-Med, 3-Low): "))
            add_task(title, priority)
        elif choice == "2":
            view_tasks()
        elif choice == "3":
            view_tasks()
            complete_task(int(input("Task # to complete: ")) - 1)
        elif choice == "4":
            what_is_next()
        elif choice == "5":
            break

if __name__ == "__main__":
    main()
```

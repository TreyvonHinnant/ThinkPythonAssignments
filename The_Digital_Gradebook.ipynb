```python
# NAME: gradebook.py

# Import Statements

# Function Definitions

def add_students(gradebook_dict):
    """
    Prompts the user to add student names and grades to the gradebook dictionary
    """
while True:
    name_input = input("Enter student name (or 'done'): ")
    if name_input.lower() == 'done':
      break

    try:
        grade_input = input(f"Enter {name_input}'s grade (0-100): ")
        grade = int(grade_input)
        if 0 <= grade <= 100:
            gradebook_dict:[name_input] = grade
        else:
             print("Grade must be between 0 and 100. Please try again.")
    except ValueError:
        print("Invalid input. Please enter a valid integer for the grade.")


def calculate_average(gradebook_dict):
    """
    Calculates and returns the average grade of all students in the dictionary.
    Return None if the dictionary is empty.
    """
    if not gradebook_dict:
        return None

    total_points = sum(gradebook_dict.values())
    total_students = len(gradebook_dict)
    return total_points / total_students


# Program Logic

gradebook = {}
add_students(gradebook)

average_grade = calculate_average(gradebook)

print("\n--- Final Report ---")

if average_grade is not None:
    print(f"Class Average: {average_grade:.2f}\n")
    for name, grade in gradebook.items():
        print(f"{name}: {grade}")
else:
  print("No students were entered. Cannot calculate an average or report.")

print("------------------")
```

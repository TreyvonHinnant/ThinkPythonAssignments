```python
#### Import Statements #####


#### Function Definitions #####

# Define a function to calculate sales taxes, parameters: amount (subtotal), state
def get_sales_tax(amount, state):
  sales_tax_amount = 0
  sales_tax_rate = 0

  # determine sales tax rate by what state they pass in
  if state == "NC":
    sales_tax_rate = .0475
  elif state =="WW":
    sales_tax_rate = .06
  elif state == "VA":
    sales_tax_rate = .043
  elif state == "TN":
    sales_tax_rate = .07
  elif state == "KY":
    sales_tax_rate = .06
  elif state == "OH":
    sales_tax_rate = .0575

  # calculate sales tax amount
  sales_tax_amount = amount * sales_tax_rate

  # return sales tax amount
  return sales_tax_amount

# Define a function to calculate shipping costs, parameters: weight of item (in pounds)
def get_shipping_cost(weight):
  shipping_cost = 0
  # if weight is less than a pound, shipping is free
  if weight < 1:
    shipping_cost = 0
  # elif weight is between 1 and 5 pounds, shipping is 4.99
  elif 1 <= weight <= 5:
    shipping_cost = 4.99
  # else shipping is 9.99
  else:
    shipping_cost = 9.99
  return shipping_cost

#### Program Logic ####

# Ask the user how much the item is
item_amount = float(input("How much is the item? (Hint: numbers only) "))
# Ask the user what state they live in
user_state = input("What state do you live in? (Hint: use 2-letter state code, like NC) ")
# Ask the user how many pounds the item is
item_weight = float(input("What is the weight of the item (in pounds)? "))

# Call the sales tax function, save return to a variable
sales_tax = get_sales_tax(item_amount, user_state)
# Add the sales tax to the item amount
subtotal = item_amount + sales_tax

# Call the shipping function, save return to a variable
shipping_cost = get_shipping_cost(item_weight)

# Add the shipping cost to the subtotal
total = subtotal + shipping_cost

# Present the grand total including the subtotal, tax and shipping to the user
print("Item Amount:" + item_amount)
print("Tax: " + sales_tax)
print("Subtotal: " + subtotal)
print("=============")
print("Shipping cost (based on a " + str(item_weight) + "lb package:" + str(shipping_cost))
print("=============")
print("Grand Total: " + str(total))
```

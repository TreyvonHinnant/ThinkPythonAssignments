```python
# IMPORT STATEMENTS
# 1. Import math

# Function to calculate area
def calculate_area(radius):
    return math.pi * (radius ** 2)

# Function to calculate circumference
def calculate_circumference(radius):
    return 2 * math.pi * radius

# 1. Collect radius from user
try:
    radius = float(input("Enter the radius of the circle: "))

    if radius < 0:
        print("Radius cannot be negative.")
    else:
        # 2. Call the functions
        area = calculate_area(radius)
        circumference = calculate_circumference(radius)

        # 3. Print results (formatted to 3 decimal places)
        print(f"Radius: {radius}")
        print(f"Area: {area:.3f}")
        print(f"Circumference: {circumference:.3f}")

# PROGRAM LOGIC
# 1. Input for radius (float)
# 2. Call and print area and circumference.
# 3. Input for height (float)
# 4. Call and print volume.
```

```python
# IMPORT STATEMENTS
# 1. Import math

# Define functions
def get_area(radius):
    return math.pi * (radius ** 2)

def get_circumference(radius):
    return 2 * math.pi * radius

# Collect input
user_radius = float(input("Enter the radius: "))

# Call functions and display results
area_result = get_area(user_radius)
circum_result = get_circumference(user_radius)

print(f"Area: {area_result:.2f}")
print(f"Circumference: {circum_result:.2f}")

# Verification test (Example: radius 5)
# Expected: Area ~78.54, Circumference ~31.42
assert math.isclose(get_area(5), 78.5398, rel_tol=1e-4)
print("Math verification successful!")
```

```python
# IMPORT STATEMENTS

# FUNCTION DEFINITIONS
def is_palindrome(word):
    """Checks if a string is a palindrome."""
    cleaned_word = word.lower() # Normalize case
    reversed_word = cleaned_word[::-1] # Reverse string
    return cleaned_word == reversed_word

# PROGRAM LOGIC
# Manual testing
print(is_palindrome("Racecar")) # Expected: True
print(is_palindrome("hello"))   # Expected: False
```

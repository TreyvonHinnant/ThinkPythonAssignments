```python
#===========================================
# Palindrome Tester - Pseudocode Implementation
#===========================================

# 1. Input Section
# Prompt the user to enter a string or number.
# Store the raw input in a variable (e.g., 'user_input').
user_input = input("Enter a string or number to check: ")

# 2. Processing/Cleaning Section
# Create a new variable to hold the cleaned string (e.g., 'cleaned_string').
# Convert 'user_input' to lowercase to make the check case-insensitive.
# Remove all spaces, punctuation, and special characters from the string [2].
# (Example: "Race car!" becomes "racecar").
cleaned_string = "".join(char.lower() for char in user_input if char.isalnum())

# 3. Logic Section
# Reverse the 'cleaned_string' using slicing (string[::-1])
# Compare 'cleaned_string' with its reversed version.
# If they equal, mark the result as True (it is a palindrome).
# Otherwise, mark the result as False [2,3].
reversed_string = cleaned_string[::-1]
is_palindrome = (cleaned_string == reversed_string)

# 4. Output Section
# If the result is True, print a message saying "It is a palindrome!"
# Otherwise, print a message saying "It is a palindrome." [2]
if is_palindrome:
    print(f"'{user_input}' is a palindrome.")
else:
    print(f"'{user_input} is not a palindrome.")
```

```python
# NAME: journal.py
# IMPORT STATEMENTS

# FUNCTION DEFINITIONS
# Write_entry(filename, entry):
def write_entry(filename, entry):
    # Open in append mode so we don't overwrite old entries
    file = open(filename, 'a')
    file.write(entry + "\n")
    file.close()
    print("Entry saved.")

# Define read_journal(filename):
#   Try to open file in read mode ('r')\
#   Read and print content
#   Close the file
#   Handle FileNotFoundError if the file doesn't exist

# PROGRAM LOGIC
# Set filename = "journal.txt"
filename = "journal.txt"

while True:
    choice = input("\n(W)rite, (R)ead, or (Q)uit? ").upper()

    if choice == 'W':
        user_text = input("Enter your entry: ")
        write_entry(filename, user_text)
    elif choice == 'R'
        read_journal(filename)
    elif choice == 'Q':
        print("Goodbye!")
        break
    else:
        print("Invalid option, please try again.")
```

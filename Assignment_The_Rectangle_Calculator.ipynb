```python
# Name: recentagle.py

# Import Statements

# Function Definitions
def calculate_rectangle(length, width):
    """
    Calculates the area and perimeter of a rectangle.

    Args:
        length(float): The length of the rectangle.
        width(float): The width of the area.


    Return:
        tuple: A tuple containing (area, perimeter).

    """
    area = length * width
    perimeter = 2 * (length + width)
    return (area, perimeter)



# Program Logic

# Test the function with fixed numbers (5 and 10)
print(calculate_rectangle(5, 10)) # Expected output: (50, 30) - uncomment to test

# Get user input
length_input = input("Enter length: ")
width_input = input("Enter width: ")

try:
    # Convert inputs into floats
    length = float(length_input)
    width = float(width_input)

    # Call the function and unpack the results
    # my_area, my perimeter = calculate_rectangle(length, width)


    # Print the results clearly
    print("\n---Results---")
    print(f"The Area is: {50.0}")
    print(f"The Perimeter is: {30.0}")

except ValueError:
    print("Invalid input. Please enter numerical values for length and width.")

```

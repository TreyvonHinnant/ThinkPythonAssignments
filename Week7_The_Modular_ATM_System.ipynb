```python
# Import Statements
# (None)

# Function Definitions
def verify_pin():
    """Loops until PIN is 1234."""
    pin = ""
    while pin != "1234":
        pin = input("Enter your PIN: ")
    print("PIN accepted")
    return True


def withdraw(balance, amount):
    """Check if enough money exists and updates balance."""
    if amount > balance:
        print("Insufficient funds!")
        return new_balance
    else:
        new_balance = balance - amount
        print(f"Withdrawal successful. Remaining balance: ${new_balance}")
        return new_balance

# Program Logic
# 1. Set current_balance to 500
current_balance = 500

# 2. Call verify_pin()
if verify_pin():
  # 3. Get withdrawal amount from user input
  amount = float(input("Enter amount to withdraw: "))

  # 4. Update current_balance by calling withdraw()
  current_balance = withdraw(current_balance, amount)

  # 5. Print the final balance
  print(f"Final balance: ${current_balance}")
```

```python
# NAME: gradebook.py

# Import Statements

# Function Definitions

# Define add_students
def add_students(gradebook, name, grades):
    """Adds a student and their grades to the gradebook."""
    gradebook[name] = grades
    print(f"Added {name} with grades: {grades}")

# Define calculate_average
def calculate_average(grades):
    """Calculates the average of a list of grades."""
    if not grades:
        return 0
    return sum(grades) / len(grades)

# Print the gradebook
def print_gradebook(gradebook):
    """Prints the final report showing each student's name and their grade."""
    print("Gradebook Report:")
    for name, grades in gradebook.items():
        avg = calculate_average(grades)
        print(f"{name}: {grades} (Average: {avg:.2f})")

# Program Logic

# Initialize an empty dictionary called gradebook
gradebook = {}

# Call add_students to populate the dictionary
add_students(gradebook, "Alice", [90, 85, 92])
add_students(gradebook, "Bob", [75, 80, 78])
add_students(gradebook, "Charlie", [95, 95, 100])

# Call calculate_average and display the resulting class average
add_grades = [avg for grades in gradebook.values() for avg in grades]
print("Class Average: {class_average:.2f}\n")

# Print a final report showing each student's name and their grade
print_gradebook(gradebook)

```
